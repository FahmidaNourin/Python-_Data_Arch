```python
def Search_Null(dataset):
  dic={}
  for col in list(dataset.columns):
    rows = []
    flage = 0
    for row in range(dataset.shape[0]):
      if str(dataset[col][row]) == "nan":
        rows.append(row)
        flage = 1
    if flage ==1:
      dic[col] = rows
  
  return dic
```

```python
Search_Null(df)
```

```python
df.shape   #all values
```

```python
df.dropna(axis = 0, how='any', inplace = True)        #droping rows which has nan values,   nan = null
```
